## 生成一个用于演示的随机数组

```python
import numpy as np
np.random.seed(61)
y = np.random.normal(size=(10, 1), loc=9.13, scale=0.4)   #  生成一个正态分布的随机数组
x = [i for i in range(1, 11)]

print(x)
print(y)
```

## 绘制一个折线图

```python
import  matplotlib.pyplot as plt
plt.plot(x, y)

plt.show()
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
plt.plot(x, y)

plt.show()
```

## 多条折线

```python
y1 = np.random.normal(size=(10, 1), loc=10, scale=0.4)
y2 = np.random.normal(size=(10, 1), loc=5, scale=2.5)
y3 = np.random.normal(size=(10, 1), loc=0, scale=1)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

## 改变坐标范围

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xlim(0, 20)
plt.ylim(-10, 100)
plt.show()
```

## 改变色彩、线型、指定坐标
更多设置参考：https://www.cnblogs.com/onemorepoint/p/7482644.html

```python
plt.plot(x, y1, '*--', label='Xiao Wang', color='olive')
plt.plot(x, y2, 'ro-', label='Caixia Ma', color='#BCEE5E')
plt.plot(x, y3, 's:', label='Xuejiam Gao', color='g')

plt.legend(framealpha=0.5, loc='lower left')  # 图例, framealpha指定透明度, loc指定位置
plt.xlabel('xxxxxx')  # 坐标轴名称
plt.ylabel('yyyyyy')  # 坐标轴名称
plt.show()
```

## 柱状图

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

y1 = [i for i in range(1, 11)]
y2 = [i for i in range(6, 16)]
y3 = [i for i in range(11, 21)]

print(x)
print(y1)
print(y2)
print(y3)
```

```python
plt.bar(x, y1, label="label1", color='green')

plt.show()
```

## 堆叠的柱状图（覆盖）

```python
plt.bar(x, y1, label="label1", color='red')
plt.bar(x, y2, label="label2", color='green')
plt.bar(x, y3, label="label3", color='blue')

plt.show()
```

```python
plt.bar(x, y1, label="label1", color='red', alpha=0.5)  # 透明度50%
plt.bar(x, y2, label="label2", color='green', alpha=0.5)
plt.bar(x, y3, label="label3", color='blue', alpha=0.5)

plt.show()
```

## 独立展示的柱状图

```python
x = np.array([i for i in range(1, 11)])
print(x)

plt.bar(x, y1, width=0.3, label="label1", color='red')
plt.bar(x+0.3, y2, width=0.3, label="label2", color='green')
plt.bar(x+0.6, y3, width=0.3, label="label3", color='blue')

plt.show()
```

```python
x1 = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
```

```python
plt.bar(x, y1, width=0.3, label="label1", color='red')
plt.bar(x+0.3, y2, width=0.3, label="label2", color='green')
plt.bar(x+0.6, y3, width=0.3, label="label3", color='blue')

plt.xticks([index + 0.3 for index in x], x1)                      # 坐标轴向右偏移0.3，并替换为自定义值
plt.grid(linestyle='--', linewidth=1, color='gray', alpha = 0.4)  # 背景网格，灰色虚线，透明度40%
plt.legend()                                                      # 展示图例
plt.show()
```

## 热力图

```python
x = np.random.rand(100).reshape(10,10)  # 生成10*10的随机数组, 范围[0,1]
```

```python
plt.imshow(x, cmap=plt.cm.hot, vmin=0, vmax=1)
plt.colorbar()  # 数值条
plt.show()
```

## 热力图会随矩阵shape自动改变

```python
x = x.reshape(20, 5)

plt.imshow(x, cmap=plt.cm.hot, vmin=0, vmax=1)
plt.colorbar()  # 数值条
plt.show()
```

## 改变热力图某点的值

```python
x[-1][-1] = 1  # 最右下角的值置为1
```

```python
plt.imshow(x, cmap=plt.cm.hot, vmin=0, vmax=1)
plt.colorbar()  # 数值条
plt.show()
```

## 图片的保存

#### 绘制一张图片，保存为png

```python
x = [i for i in range(10)]
y = [i*i for i in x]

plt.plot(x, y)
plt.savefig('test.png')
```

#### 保存为pdf

```python
from matplotlib.backends.backend_pdf import PdfPages
```

```python
x = [i for i in range(10)]
y = [i*i for i in x]

plt.plot(x, y)

pp = PdfPages('test.pdf')
pp.savefig()
plt.show()
pp.close()
plt.close()
```
